```python
import pandas as pd
from sklearn.metrics import mean_absolute_error
import seaborn as sns
sns.set()
```

```python
train = pd.read_csv('../data/train_data.csv')
test = pd.read_csv('../data/valid_data.csv')
```

```python
train = pd.get_dummies(train)
test = pd.get_dummies(test)
```

```python
X_train = train.drop('target', axis=1)
y_train = train.target

X_test = test.drop('target', axis=1)
y_test = test.target
```

```python
from sklearn.cross_decomposition import PLSRegression

pls_est = PLSRegression(3)
```

```python
pls_est.fit(X_train, y_train)
```

```python
pls_est.score(X_test, y_test)
```

```python
y_pred = pls_est.predict(X_test)
```

```python
mean_absolute_error(y_test, y_pred)
```

```python
from sklearn.svm import SVR

svr_est = SVR()
```

```python
svr_est.fit(X_train, y_train)
```

```python
svr_est.score(X_test, y_test)
```

```python
y_pred = svr_est.predict(X_test)

mean_absolute_error(y_test, y_pred)
```

```python
svr_est = SVR(kernel='linear')
```

```python
svr_est.fit(X_train, y_train)
```

```python
svr_est.score(X_test, y_test)
```

```python
y_pred = svr_est.predict(X_test)

mean_absolute_error(y_test, y_pred)
```
